```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv('C:\\Users\\hp\\OneDrive\\Desktop\\Project_internship\\2022projects\\Logistic\\customer churn predection\\churn_data.csv')
```

```python
data.dtypes
```

```python
# Drop the id column:
clean_data = data.drop('customerID', axis=1)

# Convert non-numerical values to numerical:
for col in clean_data.columns:
    if(clean_data[col].dtype == np.number):
        continue
    clean_data[col] = LabelEncoder().fit_transform(clean_data[col])

# See the data types of our columns:
print(clean_data)
```

```python
# Scale the data set:
x = clean_data.drop('Churn', axis=1)
y = clean_data['Churn']
x = StandardScaler().fit_transform(x)

# Split the data 80 training/20 testing:
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# Use the logisitic regression algorithm:
model = LogisticRegression()

# Train our model / fit our data:
model.fit(xtrain, ytrain)
```

```python
# Create predictions from the test data:
predictions = model.predict(xtest)

# See preview of predictions:
print(predictions)

# Create a report of the accuracy of our classifications:
report = classification_report(ytest, predictions)

print(report)
```
